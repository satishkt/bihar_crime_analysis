```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
from sklearn import preprocessing
```

```python
b_2010 = pd.read_excel('Data.xlsx',sheetname='2010').drop("Total")
b_2011 = pd.read_excel('Data.xlsx',sheetname='2011').drop("Total")
b_2012 = pd.read_excel('Data.xlsx',sheetname='2012').drop("Total")
b_2013 = pd.read_excel('Data.xlsx',sheetname='2013').drop("Total")
b_2014 = pd.read_excel('Data.xlsx',sheetname='2014').drop("Total")
b_2015 = pd.read_excel('Data.xlsx',sheetname='2015').drop("Total")
```

```python
def normalize(df):
    result = df.copy()
    for feature_name in df.columns:
        max_value = df[feature_name].max()
        min_value = df[feature_name].min()
        result[feature_name] = (df[feature_name] - min_value) / (max_value - min_value)
    return result
```
